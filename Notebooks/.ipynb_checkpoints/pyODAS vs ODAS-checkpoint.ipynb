# Comparing the RIS ODAS output from Matlab to pyODAS

## Start by loading the *.mat file converted with the RSI ODAS library

```python
import os 
import pyODAS.read_p
import matplotlib.pyplot as plt
import numpy as np
import scipy.io

matfile = '../TestData/UWA_013.mat'
matfile = scipy.io.loadmat(matfile)# import pyODAS.read_mat

print(matfile.keys())
```

```python
%matplotlib inline

plt.figure(figsize=(20, 5))
plt.plot(matfile['t_slow_YD'], -matfile['P_slow'])
plt.grid()
plt.ylabel('Depth (m)')
plt.xlabel('Day of {}'.format(matfile['Year'][0][0]))
plt.title('Profile depth from the RSI ODAS Matlab Library')

plt.figure(figsize=(20, 5))
plt.plot(matfile['t_slow_YD'], matfile['W_slow'])
plt.grid()
plt.ylabel('W (m/s)')
plt.xlabel('Day of {}'.format(matfile['Year'][0][0]))


n = len(matfile['sh1'])//len(matfile['t_slow_YD'])
sh1_slow = matfile['sh1'].copy()[np.arange(0, len(matfile['sh1']), n)]
sh2_slow = matfile['sh2'].copy()[np.arange(0, len(matfile['sh2']), n)]

ind = np.argsort(sh1_slow, axis=0)

vmax=0.02
vmax=0.1

# # Don't plot high values
# sh2_slow[abs(sh2_slow)>vmax] = np.nan
# sh1_slow[abs(sh1_slow)>vmax] = np.nan

# Only plot freefall
minw = 0.6
sh2_slow[matfile['W_slow']<minw] = np.nan
sh1_slow[matfile['W_slow']<minw] = np.nan


plt.figure(figsize=(20, 5))
plt.scatter(matfile['t_slow_YD'][ind], -matfile['P_slow'][ind], c=abs(sh1_slow[ind]), vmin=0, vmax=vmax)
plt.grid()
plt.ylabel('Depth (m)')
plt.xlabel('Day of {}'.format(matfile['Year'][0][0]))
plt.title('Profile depth from the RSI ODAS Matlab Library - coloured by shear')
```

## And now pyODAS
This curently generates a silly amound of screen output. Sorry about that!

```python
pfile = '../TestData/UWA_013.P'
ini_file = '../TestData/default_vehicle_attributes.ini'

output = pyODAS.read_p.main(pfile, ini_file=ini_file)
```

## Plot out some very simple comparisons


```python
print(output.keys())
```

```python
%matplotlib inline
names = ['T1_fast', 
         'T2_fast', 
         'T1_slow', 
         'T2_slow', 
         'T1', 
         'T2', 
         'T1_hres', 
         'T2_hres', 
         'gradT1', 
         'gradT2', 
         'sh1', 
         'sh2', 
         'P_slow', 
         'P_fast', 
         'Ax', 
         'Ay', 
         'V_Bat', 
         'Incl_X', 
         'Incl_Y', 
         'Incl_T', 
         'W_slow', 
         'W_fast', 
         'speed_fast', 
         'speed_slow']
# names = []
ns = 4
nr = 6
fig, axes = plt.subplots(nr, ns, figsize=(20, 20*nr/ns))
for i, name in enumerate(names):
    
    row = i//ns
    col = i-row*ns
    
    n = len(output[name]['data_physical'])
    d = output[name]['data_physical'] - matfile[name].T
    RMSE = np.sqrt(np.sum(d**2)/n)
    
    print('{} RMSE: {:.4e}'.format(name, RMSE))
          
    axes[row][col].plot(output[name]['data_physical'], matfile[name], '.')
    axes[row][col].set_title('{} ; n={} ; RMSE={:.4e}'.format(name, n, RMSE))
    
    axes[row][col].set_xlabel('pyODAS')
    axes[row][col].set_ylabel('ODAS')
    
    axes[row][col].grid()
    
```

## Converting velocity shear to real units
ODAS (and thus pyODAS) divde by the square of the fall velocity to calculate velocity shear. A cutoff (RSI say cutout - must be a Canadian thing) velocity is used to prevent very small numbers, though this cutout is still very small. Note that the speed plot below is limited to 0.05 m/s - the real speed would behave been much lower much of the time.  

To see the effect of this, convert back to 'non-physical units, and then compare.

```python
%matplotlib inline
plt.figure(figsize=(20,4))
plt.plot(matfile['t_slow_YD'], output['speed_slow']['data_physical'])
plt.plot(matfile['t_slow_YD'], matfile['speed_slow'])
plt.ylabel('Speed')
plt.grid()
plt.show()

plt.figure(figsize=(20,4))
plt.plot(matfile['t_fast_YD'], output['speed_fast']['data_physical'])
plt.plot(matfile['t_fast_YD'], matfile['speed_fast'])
plt.ylabel('Speed')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,4))

######################################
# Convert back to non-physical shear #
######################################

reconv_pyodas = output['sh1']['data_physical']*output['speed_fast']['data_physical']**2
reconv_odas = matfile['sh1']*matfile['speed_fast']**2

plt.plot(matfile['t_fast_YD'], reconv_odas, '.')
plt.plot(matfile['t_fast_YD'], reconv_pyodas, '.')

plt.ylabel('Shear x Speed squared') 
plt.title('Shear converted back to non-physical units.') 
plt.grid()

plt.show()

plt.figure(figsize=(20,4))

plt.plot(matfile['t_fast_YD'], reconv_odas[:, 0]-reconv_pyodas)

plt.ylabel('Difference in shear x Speed squared') 
plt.title('Error in shear when converted back to non-physical units.') 
plt.grid()
plt.show()
```

# Conclusion....

pyODAS is replicating the shear conversion to machine precision. When dividing each by a small number however, we get appreciable errors.
